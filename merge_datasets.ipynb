```python
panel_final = wdi.merge(
    imf,
    on=["country_code", "year"],
    how="left"
)
```

```python
missing_pct = (
    panel_final
    .isna()
    .mean()
    .mul(100)
    .round(2)
    .sort_values(ascending=False)
)

print(missing_pct)
```

```python
panel_final
```

```python
problem_countries = (
    panel_final.loc[
        panel_final.drop(
            columns=["country", "country_code", "year"],
            errors="ignore"
        ).isna().any(axis=1),
        "country_code"
    ]
    .unique()
    .tolist()
)

print(problem_countries)
```

```python
panel_final = panel_final[
    ~panel_final["country_code"].isin(problem_countries)
].copy()
```

```python
desc = panel_final.describe().T

print(desc)
```

```python
id_cols = ["country", "country_code", "year"]

features = [
    c for c in panel_final.columns
    if c not in id_cols
]
```

```python
panel_final.to_csv(
    "df_model.csv",
    index=False
)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X = scaler.fit_transform(
    panel_final[features]
)
```

```python
from sklearn.decomposition import PCA

pca = PCA()

X_pca = pca.fit_transform(X)
```

```python
import pandas as pd
import numpy as np

explained = pd.DataFrame({
    "PC": np.arange(1, len(features)+1),
    "Explained_Variance": pca.explained_variance_ratio_,
    "Cumulative": pca.explained_variance_ratio_.cumsum()
})

print(explained)
```

```python
panel_final["fiscal_risk_index"] = X_pca[:, 0]
```

```python
loadings = pd.DataFrame(
    pca.components_.T,
    index=features,
    columns=[f"PC{i+1}" for i in range(len(features))]
)

print(loadings)
```

```python
pca = PCA(n_components=7)

X_pca = pca.fit_transform(X)
```

```python
weights = pca.explained_variance_ratio_

panel_final["fiscal_risk_index"] = (
    X_pca * weights
).sum(axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 100))

panel_final["fiscal_risk_index_100"] = (
    scaler.fit_transform(
        panel_final[["fiscal_risk_index"]]
    )
)
```
