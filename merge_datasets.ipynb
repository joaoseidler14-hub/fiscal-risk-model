```python
id_cols = ["country", "country_code", "year"]

features = [
    c for c in panel_final.columns
    if c not in id_cols
]
```

```python
features
```

```python
panel_final.to_csv(
    "df_model.csv",
    index=False
)
```

PCA

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X = scaler.fit_transform(
    panel_final[features]
)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=7)

X_pca = pca.fit_transform(X)
```

```python
import pandas as pd
import numpy as np

explained = pd.DataFrame({
    "PC": np.arange(1, pca.n_components_ + 1),
    "Explained_Variance": pca.explained_variance_ratio_,
    "Cumulative": pca.explained_variance_ratio_.cumsum()
})


print(explained)
```

```python
panel_final["fiscal_risk_index"] = X_pca[:, 0]
```

```python
loadings = pd.DataFrame(
    pca.components_.T,
    index=features,
    columns=[f"PC{i+1}" for i in range(pca.n_components_)]
)

print(loadings)
```

```python
pca = PCA(n_components=6)

X_pca = pca.fit_transform(X)
```

```python

weights = (
    pca.explained_variance_ratio_
    / pca.explained_variance_ratio_.sum()
)

panel_final["fiscal_risk_index"] = (
    X_pca * weights
).sum(axis=1)
```

```python
weights
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 100))

panel_final["fiscal_risk_index_100"] = (
    scaler.fit_transform(
        panel_final[["fiscal_risk_index"]]
    )
)
```
